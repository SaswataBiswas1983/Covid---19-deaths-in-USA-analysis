### Filtering and grouping the data

```python
# Filter data for specified conditions
conditions = ["Influenza and pneumonia", "Vascular and unspecified dementia", "Diabetes", "Ischemic heart disease", "Renal failure"]
filtered_covid_data = df[df['Condition'].isin(conditions)]
```

```python
# Check the result
print(filtered_covid_data['Condition'].unique())
```

```python
# Group by state and condition, then sum the deaths across all years, months, and age groups
state_condition_data = filtered_covid_data.groupby(['State', 'Condition'])['COVID-19 Deaths'].sum().reset_index()
```

```python
# Check the result
print(state_condition_data.head())
```

```python
# Verify that the sum of COVID-19 deaths in the grouped data matches the sum in the filtered data
print(filtered_covid_data['COVID-19 Deaths'].sum())
print(state_condition_data['COVID-19 Deaths'].sum())
```

```python
# Merge with population data
merged_data = pd.merge(state_condition_data, df_population, left_on='State', right_on='State')
```

```python
# Verify the number of rows matches the expected number
print(state_condition_data['State'].nunique())
print(merged_data['State'].nunique())
```

```python
merged_data
```

```python
# Create a new column with calulations of COVID-19 death rates per 100,000 population
merged_data['Death Rate per 100,000'] = (merged_data['COVID-19 Deaths'] / merged_data['Population']) * 100000
```

```python
# Exporting "merged_data" dataframe to CSV
merged_data.to_csv(os.path.join(path, '02 Data', 'Prepared Data', 'Merge Data.csv'), index=False)
```

```python
# List of conditions
conditions = merged_data['Condition'].unique()
```

```python
# Create choropleth maps for total COVID-19 deaths
for condition in conditions:
    condition_data = merged_data[merged_data['Condition'] == condition]
    map_total_deaths = folium.Map(location=[40, -95], zoom_start=4)
    
    folium.Choropleth(
        geo_data=geojson_data_path,
        name=f'Total COVID-19 Deaths associated with {condition}',
        data=condition_data,
        columns=['State', 'COVID-19 Deaths'],
        key_on='feature.properties.name',  # Ensure this key matches the structure of your GeoJSON
        fill_color='YlOrRd',
        fill_opacity=0.7,
        line_opacity=0.2,
        legend_name=f'Total COVID-19 Deaths associated with {condition}',
        nan_fill_color='white'
    ).add_to(map_total_deaths)
    
    folium.LayerControl().add_to(map_total_deaths)
    
    file_name = os.path.join(path, '04 Analysis', 'Visualizations', f'total_deaths_{condition.replace(" ", "_").lower()}_choropleth_map.html')
    map_total_deaths.save(file_name)
    
    display(map_total_deaths)
```

## 6. Discussion
The Choropleth map shows the distribution of COVID-19 deaths across different states in the US. From map, we can observe that:

1) States with darker shades have higher numbers of COVID-19 deaths.

2) States with lighter shades have lower numbers of COVID-19 deaths.
This analysis helps us understand the geographic distribution of COVID-19 deaths and can be used to identify hotspots and areas that may need more medical resources.
